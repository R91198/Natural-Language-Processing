```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = ['the glass of milk',
             'the glass of juice',
             'the cup of tea',
             'I am a good boy',
             'I am a good developer',
             'understand the meaning of words',
             'your videos are good',]
```

```python
# Vocabulary size
vocab_size = 10000
```

```python
# One-Hot Representation
onehot_rep = [one_hot(words, vocab_size) for words in sentences]
print(onehot_rep)
```

### Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
embedded = pad_sequences(onehot_rep, padding = 'pre', maxlen = 8)
print(embedded)
```

```python
dimension = 10
```

```python
model = Sequential()
model.add(Embedding(vocab_size, dimension, input_length = 8))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded))
```
